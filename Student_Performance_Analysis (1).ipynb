# Find Average Grade

```python
df['G3'].mean()
```

# Above 15 Marks Students

```python
len(df[df['G3']>15])
```

# Correlation Check

```python
df['studytime'].corr(df['G3'])
```

# Gender Performance

```python
df.groupby('sex')['G3'].mean()
```

# Histogram

```python
plt.hist(df['G3'])
plt.title("Distribution of Grades")
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.show()
```

# Scatter Plot

```python
plt.scatter(df['studytime'], df['G3'])
plt.title("Study Time vs Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

# Bar Chart

```python
gender_avg = df.groupby('sex')['G3'].mean()
gender_avg.plot(kind='bar')
plt.title("Male vs Female Performance")
plt.show()
```
